```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=300,n_features=2,centers=4,cluster_std=1.5,random_state=42)
```

```python
kmeans=KMeans(n_clusters=4,random_state=42,n_init=10)
```

```python
kmeans.fit(x)
```

```python
kmeans.score(x)
```

```python
centers=kmeans.cluster_centers_
```

```python
centers
```

```python
labels=kmeans.labels_
```

```python
labels
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap='viridis',marker='o',edgecolors='k')
plt.scatter(centers[:,0],centers[:,1],c='red',marker='*',edgecolors='k',label='centroids')
plt.title('Kmeans')
plt.xlabel('feature1')
plt.ylabel('feature1')
plt.show()
```
