```python
# mean square error is not used when we have outliers in the data,to deal with this error,we have mean absolute error:
# for regression problems:
# 1.Mean square error
# 2 .Mean Absolute Error
# 3.Root mean square error

# Hybrid Loss Function (Huber Loss)
# Huber Loss MAE aur MSE ka combination hai.
# Chhoti errors ke liye MSE ka use karta hai (smooth optimization ke liye).
# Badi errors ke liye MAE ka use karta hai (outliers ke effect ko reduce karta hai).
# Ek delta (Î´) threshold define hota hai, jiske baad MAE apply hota hai.
```

# MSE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print("mse:",mean_squared_error(y_test['price'],pred))
```

# MAE

```python
from sklearn.metrics import mean_absolute_error
y=mean_absolute_error(y_test['price'],pred)
print("mae:",mean_absolute_error(y_test['price'],pred))
```

# RMSE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
x=mean_squared_error(y_test['price'],pred)
print("rmse:",x**0.5)
```

# Huber loss

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("huber_loss:",x+y)
```
