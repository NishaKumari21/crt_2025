```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_point=np.array([0,6])
y_point=np.array([0,250])
# created two different arrays
x_point,y_point
plt.plot(x_point,y_point)
plt.show()
# plotted a graph using x_point and y_point
```

```python
x=np.array([1,10])
y=np.array([4,300])
# created two different arrays
x,y
plt.plot(x,y,'o')
# used to get min and max point on the graph
plt.show()
```

```python
x1=np.array([1,10])
y1=np.array([2,250])
plt.plot(x1,marker='o')
# used to show min and max point with a straight line
plt.show()
```

```python
x2=np.array([1,8,4,5])
# marker='o' parameter adds circular markers at each data point.
plt.plot(x2,marker='o')
plt.show()
```

```python
x2=np.array([1,8,4,5])
# marker='*' parameter adds star markers at each data point.
plt.plot(x2,marker='*')
plt.show()
```

```python
s1=np.array([1,8,4,5])
plt.plot(s1,'o:b')
# o as a marker
# b as a color which is blue
plt.show()
```

```python
s1=np.array([1,8,4,5])
plt.plot(s1,marker='o',ms=30)
# ms:marker size
plt.show()
```

```python
s1=np.array([1,2,3,4,6])
plt.plot(s1,marker='*',ms=20)
# ms:marker size
plt.show()
```

```python
s1=np.array([1,2,3,4,6])
plt.plot(s1,marker='*',ms=20,mec='r')
# ms:marker size
plt.show()
```

```python
s3=np.array([3,4,6,7,8,9])
plt.plot(s3,marker='*',ms=30,mec='r',mfc='g')
# mec is used for boundary color
# mfc is used for the inside color of the marker
plt.show()
```

```python
s5=np.array([1,8,4,5,9])
plt.plot(s5,ls='dotted')
plt.show()
```

```python
s5=np.array([1,8,4,5,9])
plt.plot(s5,ls='dashed')
# used to chnge the line style
plt.show()
```

```python
s5=np.array([1,8,4,5,9])
plt.plot(s5,linewidth=5)
# used for the line width 
plt.show()
```

```python
a1=np.array([1,2,3,4,5,6])
a2=np.array([5,2,3,8,9,11])
plt.plot(a1)
plt.plot(a2)
plt.show()
```

```python
a1=np.array([1,2,3,4,5,6])
a2=np.array([5,2,3,8,9,11])
a3=np.array([8,9,3,11,4,5])
a4=np.array([2,3,4,7,8,9])
plt.plot(a1)
plt.plot(a2)
plt.plot(a3)
plt.plot(a4)
plt.show()
```

```python
x1=np.array([0,1,2,3,4])
x2=np.array([6,7,8,9,10])
# used for x label
plt.xlabel("x axis")
plt.ylabel("y axis")
# used for y label
plt.title("TITLE")
# used for title 
plt.plot(x1,x2)
plt.show()
```

```python
# grid 
x1=np.array([0,1,2,3,4])
x2=np.array([6,7,8,9,10])
# used for x label
plt.xlabel("x axis")
plt.ylabel("y axis")
# used for y label
plt.title("TITLE")
plt.grid()
# used for making the grid
plt.plot(x1,x2)
plt.show()
```

```python
# subplot:plot inside plot
d1=np.array([1,2,3,4,5])
d2=np.array([5,7,8,9,10])
plt.subplot(1,2,1)
plt.plot(d1,d2)
x=np.array([1,2,3,4,5,6,7])
y=np.array([10,1,3,5,6,7,4])
plt.subplot(1,2,2)
plt.plot(x,y)
```

# matplotlib scatter

```python
h1=np.array([11,12,13,14,15,16,17])
h2=np.array([3,4,5,6,7,8,9])
plt.scatter(h1,h2)
plt.show()
```

```python
# to compare the plot
g1=np.array([1,2,3,4,5,6,7])
g2=np.array([11,12,10,7,8,9,13])
g3=np.array([4,5,6,7,8,9,10])
g4=np.array([1,7,8,9,10,11,13])
plt.scatter(g1,g2)
plt.scatter(g3,g4)
plt.show()
```

```python
g1=np.array([1,2,3,4])
g2=np.array([11,12,10,15])
c=np.array(['red','green','blue','yellow'])
plt.scatter(g1,g2,c=c)
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([7,8,9,10])
c=np.array(['red','green','blue','yellow'])
size=np.array([100,200,80,40])
plt.scatter(x,y,s=size,alpha=0.5,c=c)
# alpha is used for intensity of the color
# to increase the size
plt.show()
```

```python
x=np.array([0,1,2,3])
y=np.array([7,8,9,10])
c=np.array(['red','green','blue','yellow'])
size=np.array([100,200,80,40])
plt.scatter(x,y,s=size,alpha=0.5,c=c,cmap="nipy_spectral")
# cmap="nipy_spectral" :used to create a side spectrum
# alpha is used for intensity of the color
# cmap="nipy_spectral","virdis"
# to increase the size
plt.colorbar()
plt.show()
```

# barplot

```python
x=np.array(['A','B','C','D'])
y=np.array([1,2,6,8])
plt.bar(x,y)
plt.show()
# vertical bar
```

```python
x=np.array(['A','B','C','D'])
y=np.array([1,2,6,8])
plt.barh(x,y)
plt.show()
# for horizontal bars
```

```python
x=np.array(['A','B','C','D'])
y=np.array([1,2,6,8])
plt.barh(x,y,color='red')
# color='red' used for the color of the bar
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([1,2,6,8])
plt.bar(x,y,color='red',width=0.1)
# width=0.1 used for the width of the bar
plt.show()
```

# Histogram

```python
x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

# Pie chart

```python
x=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
myexplode=[0.2,0,0,0]
plt.pie(x,labels=mylabel,startangle=90,explode=myexplode)

plt.show()
```

```python
x=np.array([1,2,3,4])
mylabel=["apple","mango","orange","banana"]
myexplode=[0.2,0,0,0]
mycolor=['red','yellow','green','blue']
plt.pie(x,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.legend(title="color")
plt.show()
```
