```python
data=sns.load_dataset("tips")
# to import sns build in dataset which is "tips"
```

```python
data.head()
```

```python
plt.plot(data['total_bill'])
plt.show()
# using matplotlib
```

```python
plt.plot(data['tip'])
plt.show()
# using matplotlib
```

```python
sns.lineplot(data['tip'])
# sns lineplot
# using sns library 
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data)
# sns scatterplot
# where x:"total_bill" and y:"tip"
```

```python
sns.histplot(data['total_bill'])
# histogram using sns library
```

```python
plt.hist(data['total_bill'])
plt.show()
# histogram using matplotlib
```

```python
sns.kdeplot(data['total_bill'])
# kde plot using sns library
```

```python
data
```

```python
# # two types of plot:
# 1.catplot:used for categorical data
# 2.relplt:eg:scatterplot
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
# relational plot
# here it is showing relation btw total_bill and tip
```

```python
data["time"].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=data)
# here relational plot is showing reation between total_bill and tip using the factor "time"
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='scatter',col='smoker',data=data)
# kind is of 2 types:
# scatter
# line
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='line',col='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',style='smoker',row='time',data=data)
```

```python
data["day"].unique()
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',hue='time',col='day',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
plt.title("Total Bill VS Tip", fontsize=15, color='green')
plt.xlabel("Total Bill", fontsize=12, color='red')
plt.ylabel("Tip", fontsize=14,color='red')
plt.show()
```
